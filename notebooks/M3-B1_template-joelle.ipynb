## Source 3 — Logs machines (texte)

```python
log_path = DATA_DIR / "logs_machines.log"
n_lines = sum(1 for _ in log_path.open())
print(f"Nombre de lignes : {n_lines:,}")
print(f"Taille fichier : {log_path.stat().st_size / 1024:.1f} Ko")

# Aperçu des 5 premières lignes
with log_path.open() as f:
    for i, line in enumerate(f):
        if i >= 5:
            break
        print(line.rstrip())
```

```python
# TODO — compter les types d'événements (INFO/WARN/ERROR), repérer les pics
```

```python
with log_path.open(encoding="utf-8") as f:
    logs = [line.rstrip() for line in f]

# Création d'un DataFrame simple
df_logs = pd.DataFrame({"raw_log": logs})

# Extraction simple des champs principaux
pattern = r"\[(.*?)\]\s+(\S+)\s+(LINE-\d+)\s+(INFO|WARN|ERROR):\s+(.*)"

df_logs[["timestamp", "site", "line_id", "level", "message"]] = df_logs["raw_log"].str.extract(pattern)

# Conversion timestamp
df_logs["timestamp"] = pd.to_datetime(df_logs["timestamp"], errors="coerce")

# Aperçu
display(df_logs.head())
```

```python
event_level_df = df_logs["level"].value_counts().reset_index()
event_level_df.columns = ["Type événement", "Nombre de lignes"]

display(event_level_df)
```

> **Observations** :
>
> - Format : Fichier texte semi-structuré avec une ligne par événement ([timestamp] site LINE-x LEVEL: message).
> - Parsing nécessaire : Oui (indispensable)
Données non directement exploitables sous forme tabulaire
Nécessite extraction via regex ou traitement string
> - Croisement avec les capteurs IoT (corrélation Roubaix line 3 ?) : TTrès pertinent. Les logs peuvent être corrélés avec les données IoT via timestamp, site et line_id. Les événements de type `ERROR` (ex. vibration_overlimit) permettent de confirmer ou d’expliquer les anomalies observées dans les capteurs, notamment sur certaines lignes (ex. Roubaix line 3). Cette complémentarité renforce la capacité du modèle à identifier des situations à risque.
